```python
mean = np.mean(df['Year_Birth'])
std = np.std(df['Year_Birth'])

lower = mean - 2 * std
upper = mean + 2 * std





mean1 = np.mean(df['MntMeatProducts'])
std1 = np.std(df['MntMeatProducts'])

lower1 = mean1 - 2 * std1
upper1 = mean1 + 2 * std1





mean2 = np.mean(df['MntGoldProds'])
std2 = np.std(df['MntGoldProds'])

lower2 = mean2 - 2 * std2
upper2 = mean2 + 2 * std2





mean3 = np.mean(df['NumDealsPurchases'])
std3 = np.std(df['NumDealsPurchases'])

lower3 = mean3 - 2 * std3
upper3 = mean3 + 2 * std3





mean4 = np.mean(df['NumWebPurchases'])
std4 = np.std(df['NumWebPurchases'])

lower4 = mean4 - 2 * std4
upper4 = mean4 + 2 * std4





mean5 = np.mean(df['NumCatalogPurchases'])
std5 = np.std(df['NumCatalogPurchases'])

lower5 = mean5 - 2 * std5
upper5 = mean5 + 2 * std5





mean6 = np.mean(df['NumWebVisitsMonth'])
std6 = np.std(df['NumWebVisitsMonth'])

lower6 = mean6 - 2 * std6
upper6 = mean6 + 2 * std6

df2 = df[ (df['Year_Birth'] >= lower) & (df['Year_Birth'] <= upper) &
            (df['MntMeatProducts'] >= lower1) & (df['MntMeatProducts'] <= upper1) &
              (df['MntGoldProds'] >= lower2) & (df['MntGoldProds'] <= upper2) &
                (df['NumDealsPurchases'] >= lower3) & (df['NumDealsPurchases'] <= upper3) &
                  (df['NumWebPurchases'] >= lower4) & (df['NumWebPurchases'] <= upper4) &
                    (df['NumCatalogPurchases'] >= lower5) & (df['NumCatalogPurchases'] <= upper5) &
                      (df['NumWebVisitsMonth'] >= lower6) & (df['NumWebVisitsMonth'] <= upper6)]
```

```python
num_col = [col for col in df2.columns if df2[col].dtype != 'object']


fig,axs = plt.subplots(len(num_col),1,figsize = (7,50))

for i,j in zip(num_col,range(len(num_col))):
  plt.title(i)
  axs[j] = plt.boxplot(df2[i],vert = False)
  plt.show()
```

```python
cat_col = [col for col in df2.columns if df2[col].dtype == 'object']

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in cat_col:
  df2[col] = encoder.fit_transform(df2[col])

df2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled = scaler.fit_transform(df2)

scaled_pd = pd.DataFrame(scaled,columns = df2.columns)
```

```python
scaled_pd.head()
```

```python
X = scaled_pd
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)

print(silhouette_score(X,kmeans.labels_))
```

```python
from sklearn.cluster import AgglomerativeClustering

agg_cluster = AgglomerativeClustering(n_clusters = 4)

y = agg_cluster.fit_predict(X)
```

```python
print(silhouette_score(X,y))
```
