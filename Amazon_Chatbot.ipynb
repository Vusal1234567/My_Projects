```python
from langchain_openai import ChatOpenAI
```

```python
!pip install langchain_community

from langchain_community.llms import HuggingFaceHub
import os

hf = HuggingFaceHub(
    repo_id = 'meta-llama/Llama-3.2-1B',
    model_kwargs = {'temperature' : 0.6,'max_length' : 100}
)
```

```python
!pip install sentence_transformers

from langchain.embeddings import HuggingFaceEmbeddings
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.amazon.com/")

docs = loader.load()
```

```python
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter()

documents = text_splitter.split_documents(docs)
```

```python
!pip install faiss-cpu

from langchain_community.vectorstores import FAISS

vector = FAISS.from_documents(documents,embeddings)
```

```python
retriever = vector.as_retriever()
```

```python
retriever.get_relevant_documents('What is Amazon')[0]
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import MessagesPlaceholder

prompt = ChatPromptTemplate([
    MessagesPlaceholder(variable_name = 'chat_history'),
    ('user','{input}'),
    ('user','Generate a search query relevant to the conversation')
])
```

```python
from langchain.chains import create_history_aware_retriever

retriever_chain = create_history_aware_retriever(hf,retriever,prompt)
```

```python
from langchain_core.messages import HumanMessage,AIMessage

chat_history = [
    HumanMessage(
        content = 'Is Amazon big company?'
    ),
    AIMessage(
        content = 'Yes'
    )
]
```

```python
result = retriever_chain.invoke({
    'chat_history' : chat_history,
    'input' : 'Tell me how?'
})

print(result)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system","Answer the user's questions based on the below\
 context: \n\n{context}"),
    MessagesPlaceholder(variable_name = 'chat_history'),
    ('user','{input}')
])
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

documents_chain = create_stuff_documents_chain(hf,prompt)
```

```python
from langchain.chains import create_retrieval_chain

retriever_chain = create_retrieval_chain(retriever_chain,documents_chain,)
```

```python
chat_history = [
    HumanMessage(
        content = 'Is Amazon big company?'
    ),
    AIMessage(
        content = 'Yes'
    )
]
```

```python
while True:

  user_input = input("Give Your Question, for leaving chat press 'q': ")

  if user_input == 'q':
    break

  response1 = retriever_chain.invoke({
    'chat_history' : chat_history,
    'input' : user_input
  })

  print(response1['answer'])


  chat_history = [
    HumanMessage(
        content = user_input
    ),
    AIMessage(
        content = response1['answer']
    )
]
```
