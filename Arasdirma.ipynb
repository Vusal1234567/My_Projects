ARASDIRMA - TRANSFORMERS(ENCODER ; DECODER ; Sequence-to-Sequence)

Transformerlər 2017-ci ildə Vaswani və digərləri tərəfindən təqdim edilən və "Attention is All You Need" adlı məqalədə təsvir edilən neyron şəbəkə arxitekturasıdır. Transformerlər təkrarlanan neyron şəbəkələr (RNN) və konvolyusiya neyron şəbəkələri (CNN) kimi ənənəvi modellərdən fərqli olaraq, tamamilə self-attention mexanizminə əsaslanır.

image.png

Gorduyumuz kimi Transformerler 2 esas hisseden ibaretdir: Encoder ve Decoder.

Encoder input olaraq metn goturur ve masinin oyrenmesi ucun hemin metni vektorlara cevirir.

Decoder Encoder'in outputunu ve diger inputlari qebul edir ve onlari target generate etmek ucun istifade edir.

Encoder ayri bir sekilde istifade oluna biler,bu zaman ancaq metn inputlari vektorlara cevirir.Meselen,cumle 3 sozden ibaretdir "Bakiya xos geldiz",herbir sozu vektora cevirdikde diger sozlerinde tesiri olur.Meselen "xos" sozunu vektora cevirdikde "Bakiya" ve "geldiz" sozlerinden tesirlenir.Encoder sentence classification(cumle təsnifatı) kimi tapsiriqlarda istifade olunur.Encoder ucun istifade ede bileceyemiz modeller:ALBERT,BERT,DistilBERT,ELECTRA,RoBERTa.

Decoder ayri bir sekilde istifade oluna biler,bu zaman input-a sozler gelir,bu sozler vektorlara cevrilir ve gelecek sozler texmin olunmaga calisilir.Amma burda sozu vektora cevirerken yalniz evvelki sozlerden tesirlenir,cunki gelecek sozlere birnov access-i yoxdu onlari generate etmelidir.Meselen "Bakiya xos geldiz" input kimi gelirse,'xos' sozunu vektora cevirdikde yalniz 'Bakiya' sozune baxilir.Herbirini vektora cevirdikde sonra ise yeni text generate etmeye calisir,meselen "Bakiya xos geldiz,hemise siz gelin".Decoder text generation kimi tapsiriqlarda istifade olunur.Decoder ucun istifade ede bileceyemiz modeller:CTRL,GPT,GPT-2,Transformer XL.

Sequence-to-sequence modelleri hem encoder hemde decoder istifade edir.Encoder input olaraq metn goturur ve masinin oyrenmesi ucun hemin metni vektorlara cevirir.Decoder Encoder'in outputunu ve diger inputlari qebul edir ve onlari target generate etmek ucun istifade edir.Encoder-de attention layerlerin butun sozlere access-i var,lakin Decoder-de attention layerlerin verilen sozden evvel yerlesen sozlere access-i var.Sequence-to-sequence modelleri summarization, translation, veya generative question answering kimi tapsiriqlarda istifade olunur.Sequence-to-sequence modelleri ucun istifade ede bileceyemiz modeller:BART,mBART,Marian,T5.
